```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# 線分の傾きと切片
a1 = (5 - 1)/(6 - 0)
b1 = 1

# 中点
cx = (0 + 6) / 2
cy = (1 + 5) / 2

# 直行する線分の傾き
a2 = -1 / a1

# 直行する線分の切片
b2 = cy - a2 * cx

# 直線の式
x = np.arange(0, 7)
y1 = a1*x + b1
y2 = a2*x + b2

# 描画
plt.plot(x, y1)
plt.plot(x, y2)
plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# 三角関数で円

# 角度
th = np.arange(0, 360)

# 円周上の点Pの座標
x = np.cos(np.radians(th))
y = np.sin(np.radians(th))

# 描画
plt.plot(x, y)
plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# 角度
th = np.arange(0, 360)

# 円周上の点Pの座標
r = 5
x = r * np.cos(np.radians(th)) + 2
y = r * np.sin(np.radians(th)) + 3


# 描画
plt.plot(x, y)
plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```

```python
# arctangent
import numpy as np

rad = np.arctan2(3, 4) # 角度を求める
th = np.degrees(rad) # 弧度法から度数法へ
print(th)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# 円の中心
a = 200
b = 300

# 円の方程式
r = 300
x = np.arange(a-r, a+r+1)

y = np.sqrt(r**2 - (x-a)**2) + b
y2 = -y + 2*b

plt.plot(x, y)
plt.plot(x, y2)
plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```
